# Storytelling Data Visualization on Exchange Rates

## Dataset

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

This dataset has 41 columns and 6456 rows with none of them being null. Most columns are objects, but a few are floats.

## Data Cleaning
Next we will clean the data to make it easier to work with for the purposes examining the euro to US dollar exchange rate.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'}, 
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

We can see `US_dollar` has 62 rows with a dash instead of a numeric value. Let's fix this.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar.dtypes
```

`Us_dollar` also appears to be an object, so let's conver it to a float.

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.dtypes
```

## Rolling Mean
Next, we will create a line plot to visualize how the euro to dollar exchange rate has changed over time.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

We can this graph is not very smooth. This is because the exchange rate changes each day, creating many variations over multiple years. This makes the graph unpleasant to look at and more difficult to read. We can solve this by using a rolling mean. This will simplifiy the graph, by taking the average of the values over set periods of time (i.e. the rolling window), leading to less data points and a smoother graph.

Let's see what the rolling mean looks like with a 7 day, 30 day, 50 day, 100 day, and 365 day rolling windows.

```python
plt.figure(figsize=(10, 7))

plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6], [7, 30, 50, 100, 365]):
    plt.subplot(3, 2, i)
    plt.plot(euro_to_dollar['Time'],
            euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window: ' + str(rolling_mean), weight='bold')

plt.tight_layout()
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## Storytelling Data Visualization Examples
Here are a couple story ideas for our data:
* How the euro-dollar rate has changed during the 2007-2008's financial crisis.
* How the euro-dollar rate changed under the last three US presidents: George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021).

### Financial Crisis

```python
financial_crisis = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2006
                   ) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >= 2007
                   ) & (euro_to_dollar.Time.dt.year <= 2008)]

### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth=1, color='#A6D785')

### Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='#e23d28')

### Highlihting the peak of the crisis
ax.axvspan(xmin=13955, xmax=14130, ymin=0.09,
           alpha=0.3, color='grey')

### Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 13108
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365

x_start = 13005
y = 1.193   
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(x_start, y, rate, alpha=0.5, fontsize=11)
    y += 0.1

## Adding a title and a subtitle    
ax.text(x_start, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold')
ax.text(x_start, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)

### Adding a signature
ax.text(x_start, 1.07, '©DATAQUEST' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

### Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```
